```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name_fr_en = "Helsinki-NLP/opus-mt-fr-en"
tokenizer_fr_en = AutoTokenizer.from_pretrained(model_name_fr_en)
model_fr_en = AutoModelForSeq2SeqLM.from_pretrained(model_name_fr_en)

text = "Ce cours est produit par Hugging Face."

# Prepare the input for the model
input_ids = tokenizer_fr_en(text, return_tensors="pt").input_ids

# Generate translation
outputs = model_fr_en.generate(input_ids)

# Decode the generated tokens
translation_text = tokenizer_fr_en.decode(outputs[0], skip_special_tokens=True)
print(translation_text)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name_en_hi = "Helsinki-NLP/opus-mt-en-hi"
tokenizer_en_hi = AutoTokenizer.from_pretrained(model_name_en_hi)
model_en_hi = AutoModelForSeq2SeqLM.from_pretrained(model_name_en_hi)

english_text = "My name is Bhavesh Negi"

# Prepare the input for the model
input_ids = tokenizer_en_hi(english_text, return_tensors="pt").input_ids

# Generate translation
outputs = model_en_hi.generate(input_ids)

# Decode the generated tokens
hindi_translation = tokenizer_en_hi.decode(outputs[0], skip_special_tokens=True)
print(hindi_translation)
```

```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
from transformers import AutoModel

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(checkpoint)
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
```

```python
print(outputs.logits.shape)
```

```python
print(outputs.logits)
```

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

tokens = tokenizer.tokenize(sequence)
ids = tokenizer.convert_tokens_to_ids(tokens)

input_ids = torch.tensor([ids])
print("Input IDs:", input_ids)

output = model(input_ids)
print("Logits:", output.logits)
```

```python
batched_ids = [ids, ids]
input_ids = torch.tensor(batched_ids)
print("Input IDs:", input_ids)

output = model(input_ids)
print("Logits:", output.logits)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

sequence1_ids = [[200, 200, 200]]
sequence2_ids = [[200, 200]]
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]

print(model(torch.tensor(sequence1_ids)).logits)
print(model(torch.tensor(sequence2_ids)).logits)
print(model(torch.tensor(batched_ids)).logits)
```

```python
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]

attention_mask = [
    [1, 1, 1],
    [1, 1, 0],
]

outputs = model(torch.tensor(batched_ids), attention_mask=torch.tensor(attention_mask))
print(outputs.logits)
```
